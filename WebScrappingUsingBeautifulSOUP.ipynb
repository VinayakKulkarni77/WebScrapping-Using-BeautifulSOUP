### Empty list to store all data about a product with different features

```python
products=[]              #List to store the name of the product
prices=[]                #List to store price of the product
ratings=[]               #List to store rating of the product
RAM_ROM = []                #List to store RAM & ROM Specifications                
display = []                  #List to store display
camera = []                  #List to store camera
Battery = []               #List to store Battery
warrenty=[]                #List to store warrenty
```

### Links of a page to fetch data for different brand and multiple pages

```python
link1="https://www.flipkart.com/search?sid=tyy%2F4io&sort=recency_desc&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIkxhdGVzdCBTYW1zdW5nIG1vYmlsZXMgIl0sInZhbHVlVHlwZSI6Ik1VTFRJX1ZBTFVFRCJ9fX19fQ%3D%3D&wid=1.productCard.PMU_V2_1&p%5B%5D=facets.brand%255B%255D%3D"
link2="&page="
```

### Product brand names 

```python
BrandNames=['realme','Infinix','Samsung','POCO','Mi','Apple','ASUS','Honor','LAVA','Micromax','OPPO','Vivo','OnePlus']
```

### To fetch the product details by brand and also for multiple pages with different specifications

```python
for i in BrandNames:
    for page_no in range(1,5):
        req=requests.get(link1+i+link2+str(page_no))
        soup=BeautifulSoup(page.content,"html.parser")
        for data in soup.findAll('div',class_='_3pLy-c row'):
            names=data.find('div',attrs={'class':'_4rR01T'})
            price=data.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
            rating=data.find('div',attrs={'class':'_3LWZlK'})
            specification=data.find('div',attrs={'class':'fMghEO'})
            for each in specification:
                col=each.find_all('li',attrs={'class':'rgWa7D'})
                RAM_=col[0].text
                display_=col[1].text
                camera_=col[2].text
                Battery_=col[3].text
                warrenty_=col[4].text
            products.append(names.text)
            prices.append(price.text)
            RAM_ROM.append(RAM_)
            display.append(display_)
            camera.append(camera_)
            Battery.append(Battery_)
            warrenty.append(warrenty_)
            ratings.append(rating)    
```

### printing the length of list

```python
print(len(products))
print(len(ratings))
print(len(prices))
print(len(RAM_ROM))
print(len(display))
print(len(camera))
print(len(Battery))
print(len(warrenty))
```
